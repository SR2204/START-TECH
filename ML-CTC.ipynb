```python
data.columns =['S.No.', 'College', 'Role', 'City Type', 'Previous CTC', 'Previous job changes', 'Graduation marks', 'Exp(months)', 'CTC']
data.head(10)
```

```python
#Replacing NAN as 0, because many times checking for null values doesn't detect NAN as null.

zeroes=['College','Role','City Type','Previous CTC','Previous job changes','Graduation marks','Exp(months)','CTC']
for i in zeroes:
    data[i].replace(0,np.nan,inplace=True)
    
data.head(10)
```
